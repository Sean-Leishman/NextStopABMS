```python
import pandas as pd 
import numpy as np
import math
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
df = pd.read_csv("stations.csv")
```

```python
def fix_string(line):
    if len(line) == 3:
        if line[2] == "C":
            return line[:2] + "00"
        return line[:2] + "0" + line[2:]
    else:
        return line
```

```python
df['STN_NO'] = df['STN_NO'].apply(lambda x: fix_string(x))
df['LINE'] = df['STN_NO'].str[:2]
df['NUMBER'] = df['STN_NO'].str[2:]
    
```

```python
ss = StandardScaler()
df['X'] = np.floor(ss.fit_transform(df[['X']]) * 20)
df['Y'] = np.floor(ss.transform(df[['Y']]) * 25) - 15
```

```python
df = df.fillna(-0.5)
```

```python
df[df['WEEKDAY_SAT_INC_START_TIME'] == "FALSE"]
```

```python
df['SUN_INC_START_TIME'] = df['SUN_INC_START_TIME'].astype('int') * 2
df['INC_END_TIME'] = df['INC_END_TIME'].astype('int') * 2
df['WEEKDAY_SAT_DEC_START_TIME'] = df['WEEKDAY_SAT_DEC_START_TIME'].astype('int') * 2
df['WEEKDAY_SAT_INC_START_TIME'] = df['WEEKDAY_SAT_INC_START_TIME'].astype('int') * 2
df['SUN_DEC_START_TIME'] = df['SUN_DEC_START_TIME'].astype('int') * 2
df['DEC_END_TIME'] = df['DEC_END_TIME'].astype('int') * 2
```

```python
connections = df.groupby("STN_NAME").min()
df['CONNECTION_ID'] = df['STN_NAME'].apply(lambda x: connections.loc[x]["OBJECTID"])
```

```python
df = df.astype({'X':int, 'Y': int})
```

```python
def nextId(row):
    a = df.loc[(df.LINE == row.LINE) & (df.NUMBER > row.NUMBER)]['NUMBER'].min()
    return a
```

```python
def prevId(row):
    a = df.loc[(df.LINE == row.LINE) & (df.NUMBER < row.NUMBER)]['NUMBER'].max()
    return a
```

```python
df = df.drop(df.loc[(df.STN_NO == "CC18")].index)
```

```python
df['NEXT_STATION_NUMBER'] = df.apply(lambda x: nextId(x) , axis=1)
df['PREV_STATION_NUMBER'] = df.apply(lambda x: prevId(x) , axis=1)
```

```python
df.loc[df.LINE == "CC"]
```

```python
df = df.fillna(-1)
```

```python
df3 = pd.read_csv("transport_node_train_202302.csv")

df3['MAX_CAPACITY'] = df3['TOTAL_TAP_IN_VOLUME'] + df3["TOTAL_TAP_OUT_VOLUME"]
df3['MIN_CAPACITY'] = df3['TOTAL_TAP_IN_VOLUME'] + df3["TOTAL_TAP_OUT_VOLUME"]
max_capacity = df3.groupby(["PT_CODE"])[['MAX_CAPACITY']].max() / 28
min_capacity = df3.groupby(["PT_CODE"])[['MIN_CAPACITY']].min() / 28 
df3 = max_capacity.join(min_capacity)
df3 = df3.reset_index()
df3['PT_CODE'] = df3['PT_CODE'].apply(lambda x: x.split("/"))
df3 = df3.explode('PT_CODE')
```

```python
df3['PT_CODE'] = df3['PT_CODE'].apply(lambda x: fix_string(x))
df3 = df3.rename({'PT_CODE':'STN_NO'}, axis=1)
```

```python
df = df.merge(df3, on="STN_NO")
```

```python
df.to_csv("formatted_stations.csv")
```
