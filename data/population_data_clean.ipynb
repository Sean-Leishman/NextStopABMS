```python
# Table Title: Resident Population by Planning Area/Subzone of Residence, Age Group and Sex (Census of Population 2020)
df0 = pd.read_csv("17560-3.csv", skiprows=11, skipfooter=22, index_col=0).iloc[:, :20]
df0 = df0[df0.index.str.contains(" - Total")].replace("-", 0)
df0.index = [x.replace(" - Total", "").upper() for x in df0.index]
df0 = df0.astype("int")
total = df0['Total'].sum()
df0.head()
```

```python
df0 = df0[['Total']].join(df0.loc[:, df0.columns != "Total"].apply(lambda x: x / df0.loc[x.name, "Total"], axis=1).cumsum(axis=1))
df0['Total'] = df0['Total'] / total 
```

```python
df0 = df0 / df0['Total'].sum()
```

```python
df0.to_csv("total_resident_population.csv", header=None)
```

```python
# Table Title: Employed Residents Aged 15 Years and Over by Planning Area of Workplace, Usual Mode of Transport to Work and Planning Region of Residence (Census of Population 2020)
df1 = pd.read_csv("17586-0.csv", index_col=0,skiprows=9, skipfooter=18, header=[0,1])
df1 = df1.replace("-", 0)
df1 = df1.astype("int")
df1 = df1 / total
```
